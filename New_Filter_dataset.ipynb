```python
# Список колонок для удаления
columns_to_remove = [
    '_id', 'TimeStamp', 'PART_FACT_PLAN_DATE', 'EQUIP_CD', 'EQUIP_NAME', 
    'Reason', 'Injection_Time', 'Filling_Time', 'Plasticizing_Time',
    'Clamp_Close_Time', 'Cushion_Position', 'Switch_Over_Position', 
    'Plasticizing_Position', 'Clamp_Open_Position', 'Max_Screw_RPM', 
    'Average_Screw_RPM', 'Max_Injection_Pressure', 'Max_Switch_Over_Pressure',
    'Average_Back_Pressure'
]
```

```python
df_cleaned = df.drop(columns=columns_to_remove)
df_cleaned.to_csv("Dataset_cleaned.csv", index=False)
```

```python
print("\nИнформация о новом датасете:")
print(f"Количество строк: {len(df_cleaned)}")
print(f"Количество колонок: {len(df_cleaned.columns)}")
print("\nСписок оставшихся колонок:")
for col in df_cleaned.columns:
    print(f"- {col}")
```

```python
# Загружаем очищенный датасет
print("Загрузка очищенного датасета...")
df = pd.read_csv('Dataset_cleaned.csv')
```

```python
# Фильтруем данные по PART_NAME
print("Фильтрация данных по PART_NAME...")
filtered_df = df[df['PART_NAME'].isin(["RG3 MOLD'G W/SHLD, RH"])]
```

```python
# Сохраняем отфильтрованный датасет
print("Сохранение отфильтрованного датасета...")
filtered_df.to_csv('Dataset_filtered.csv', index=False)
```

```python
# Выводим информацию о датасете
print("\nИнформация о отфильтрованном датасете:")
print(f"Количество строк: {len(filtered_df)}")
print(f"Количество колонок: {len(filtered_df.columns)}")
print("\nСписок колонок:")
for col in filtered_df.columns:
    print(f"- {col}")

# Выводим статистику по числовым колонкам
print("\nСтатистика по числовым колонкам:")
print(filtered_df.describe()) 
```
