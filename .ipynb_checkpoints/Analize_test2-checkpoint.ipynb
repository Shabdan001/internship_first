```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from datetime import datetime
import matplotlib.dates as mdates
from matplotlib.gridspec import GridSpec
```

```python
df = pd.read_csv("../first_project/Dataset.csv")
df.head(5)
```

```python
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])
df['Reason'] = df['Reason'].replace({
    '가스': 'gas',
    '초기허용불량': 'Initial_allowable_defect',
    '미성형': 'Short_plastic_surgery'
})
```

```python
# Base info about dataset 
print("\n About Dataset")
print(df.info())

print("\nBase statistics")
print(df.describe()) 

# Missing value check 
print("\nMissing value:")
print(df.isnull().sum())
```

```python
# Time series analize
numeric_columns = df.select_dtypes(include=[np.number]).columns

# groupping and mean ident 
daily_metrics = df.groupby(df['TimeStamp'].dt.date).agg({
    col: 'mean' for col in numeric_columns
})

daily = df.groupby(df['TimeStamp'].dt.date).agg({
    'Injection_Time': 'mean', 
    'Cycle_Time': 'mean'
})
```

```python
# passRate create 
daily_metrics['PassRate'] = df.groupby(df['TimeStamp'].dt.date)['PassOrFail'].apply(lambda x: (x == 'Y').mean())
```

```python
# Reason analize 
print("\nReason...")
reason_counts = df['Reason'].value_counts()
print('\nReason gropping:')
print(reason_counts)
```

```python
# We create graphs for each numerical column
n_cols = 2  #Number of columns in the graph grid
n_rows = (len(numeric_columns) + 1) // n_cols + 1  # +1 for the passage schedule

plt.figure(figsize=(15, 5*n_rows))

# Graphs for numeric columns
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.plot(daily_metrics.index, daily_metrics[col])
    plt.title(f'{col} to day')
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.grid(True)

    # Setting the date format
    ax = plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m'))
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=max(1, len(daily_metrics)//10)))
    plt.xticks(rotation=45)



# Pass percentage chart
plt.subplot(n_rows, n_cols, len(numeric_columns) + 1)
plt.plot(daily_metrics.index, daily_metrics['PassRate'])
plt.title('Percentage of passing control by day')
plt.xlabel('Date')
plt.ylabel('Percentage of passing')
plt.grid(True)
# Setting the date format
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=max(1, len(daily_metrics)//10)))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 2. Причины брака
reasons = df.groupby([df['TimeStamp'].dt.date, 'Reason']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
for reason in reasons.columns:
    plt.plot(reasons.index, reasons[reason], '-', label=reason)
plt.title('Причины брака')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Корреляционная матрица
corr_cols = [
    'Reason', 'Injection_Time', 'Filling_Time', 'Plasticizing_Time', 'Cycle_Time',
    'TimeStamp', 'PART_NAME', 'EQUIP_CD', 'EQUIP_NAME',
    'PassOrFail', 'Barrel_Temperature_1', 'Barrel_Temperature_2', 
    'Barrel_Temperature_3', 'Barrel_Temperature_4', 'Barrel_Temperature_5',
    'Mold_Temperature_3'
]
```

```python
# График распределения причин брака
plt.figure(figsize=(12, 6))
sns.barplot(x=reason_counts.index, y=reason_counts.values)
plt.title('Распределение причин брака')
plt.xlabel('Причина')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Создаем копию данных для корреляционного анализа
corr_data = df[corr_cols].copy()
```

```python
# Function for defect analysis
def analyze_defects(start_date, end_date, title_suffix):
    # Filter data by period
    mask = (df['TimeStamp'] >= start_date) & (df['TimeStamp'] <= end_date)
    filtered_df = df[mask]
    
    # Get list of all defect types
    defect_types = filtered_df['Reason'].unique()
    
    # Create complex plot
    fig = plt.figure(figsize=(15, 10))
    gs = GridSpec(2, 2, figure=fig)
    
    # 1. Line plot for each defect type (hourly)
    ax1 = fig.add_subplot(gs[0, :])
    for defect_type in defect_types:
        # Group by hour and count defects
        hourly_data = filtered_df[filtered_df['Reason'] == defect_type].copy()
        hourly_data['date'] = hourly_data['TimeStamp'].dt.date
        hourly_data['hour'] = hourly_data['TimeStamp'].dt.hour
        
        hourly_defects = hourly_data.groupby(['date', 'hour']).size().reset_index()
        
        # Create datetime for plotting
        hourly_defects['plot_time'] = pd.to_datetime(
            hourly_defects['date'].astype(str) + ' ' + 
            hourly_defects['hour'].astype(str) + ':00'
        )
        
        # Plot only if there are defects
        if not hourly_defects.empty:
            ax1.plot(hourly_defects['plot_time'], hourly_defects[0], 
                    marker='o', linewidth=2, markersize=8, label=defect_type)
    
    ax1.set_title(f'Defect Dynamics by Type ({title_suffix})', fontsize=14)
    ax1.set_xlabel('Date and Time', fontsize=12)
    ax1.set_ylabel('Number of Defects', fontsize=12)
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend()
    
    # 2. Bar chart of total defects by type
    ax2 = fig.add_subplot(gs[1, 0])
    total_defects = filtered_df.groupby('Reason').size()
    sns.barplot(x=total_defects.index, y=total_defects.values, ax=ax2)
    ax2.set_title('Total Defects by Type', fontsize=14)
    ax2.set_xlabel('Defect Type', fontsize=12)
    ax2.set_ylabel('Count', fontsize=12)
    ax2.tick_params(axis='x', rotation=45)
    
    # Add values on bars
    for i, v in enumerate(total_defects.values):
        ax2.text(i, v, str(v), ha='center', va='bottom', fontsize=10)
    
    # 3. Pie chart of defect distribution
    ax3 = fig.add_subplot(gs[1, 1])
    ax3.pie(total_defects.values, labels=total_defects.index, autopct='%1.1f%%',
            startangle=90, shadow=True, explode=[0.1]*len(total_defects))
    ax3.set_title('Defect Type Distribution', fontsize=14)
    
    plt.tight_layout()
    plt.show()
    
    # Print statistics
    print(f"\nDefect Statistics for period {title_suffix}:")
    for defect_type in defect_types:
        # Hourly statistics
        hourly_data = filtered_df[filtered_df['Reason'] == defect_type].copy()
        hourly_data['date'] = hourly_data['TimeStamp'].dt.date
        hourly_data['hour'] = hourly_data['TimeStamp'].dt.hour
        
        hourly_stats = hourly_data.groupby(['date', 'hour']).size()
        
        if not hourly_stats.empty:
            print(f"\nDefect Type: '{defect_type}'")
            print(f"Total Count: {hourly_stats.sum()}")
            print(f"Average per hour: {hourly_stats.mean():.2f}")
            print(f"Maximum per hour: {hourly_stats.max()}")
            print(f"Minimum per hour: {hourly_stats.min()}")
            print(f"Hour with maximum: {hourly_stats.idxmax()}")
            print(f"Hour with minimum: {hourly_stats.idxmin()}")
            print("-" * 50)
```

```python
# Analyze period 26-29 October
analyze_defects('2020-10-26', '2020-10-29', '26-29 October 2020')

# Analyze period 26-29 October
analyze_defects('2020-10-21', '2020-11-26', '10.21-11.26 October 2020')
```

```python
# Преобразуем категориальные колонки в числовой формат
categorical_cols = ['Reason', 'PART_NAME', 'EQUIP_CD', 'EQUIP_NAME', 'PassOrFail']
for col in categorical_cols:
    if col in corr_data.columns:
        corr_data[col] = corr_data[col].astype('category').cat.codes
```

```python
# Преобразуем временные метки в числовой формат
time_cols = ['TimeStamp', 'PART_FACT_PLAN_DATE']
for col in time_cols:
    if col in corr_data.columns:
        # Преобразуем Unix timestamp в datetime
        corr_data[col] = pd.to_datetime(corr_data[col], unit='ns').astype('int64')
```

```python
plt.figure(figsize=[15, 12])
sns.heatmap(corr_data.corr(), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title("Corr_matrix")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
